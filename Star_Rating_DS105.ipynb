```python
#importing libraries
import pandas as pd
from plotnine import *
from google.colab import files 

# enabling the upload of cleaned data csv to colab (not needed if colab isn't being used)
uploaded = files.upload()

# reading the csv
df = pd.read_csv('amazon_reviews.csv')

# storing the first column in the dataframe
first_column = df.columns[2]

# grouping data in the first column and calculating frequency
grouped = df[first_column].value_counts().reset_index(name='frequency')
grouped = grouped.rename(columns={'index': first_column})
print(grouped)
# plotting the frequency as a bar chart with axis labelled,title,theme and axis parameters
p = ggplot(grouped, aes(x=first_column, y='frequency')) + geom_bar(stat='identity') + theme_classic(base_size=14, base_family='Arial')+ ggtitle("Bar Chart showing spread of Star Rating")
p = p + scale_y_continuous(limits=(0, 2000))
# printing the bar chart
print(p)
```
