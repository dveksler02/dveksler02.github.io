```python
# importing libraries
import pandas as pd
from plotnine import*

# reading the csv
df = pd.read_csv('amazon_reviews.csv')

# getting name of last column
date_column = df.columns[-1]

# converting date column to a datetime format
df[date_column] = pd.to_datetime(df[date_column])

# extracting month from the date column
df['month'] = df[date_column].dt.month

# grouping by month and calculating frequency
grouped = df.groupby(['month']).size().reset_index(name='frequency')

# plottin frequency against the month as a line graph adding axis,titles,themes and scale parameters for the x and y axis
p = ggplot(grouped, aes(x='month', y='frequency')) + geom_line() + ggtitle("Frequency of Reviews from Dec 7th 2022")+ theme_classic(base_size=14, base_family='Arial')
p = p + scale_x_continuous(limits=(0, 12),breaks=range(0, 13))

# printing the line graph
print(p)

```
