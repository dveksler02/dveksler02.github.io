```python
# importing libraries
import pandas as pd
from plotnine import *
from google.colab import files 

# enabling the upload of cleaned data csv to colab (not needed if colab isn't being used)
uploaded = files.upload()

# reading the csv
df = pd.read_csv('amazon_reviews.csv')

# calculate the length of each text in the 3rd column
lengths = [len(text) for text in df.iloc[:, 1]]

# creating a new dataframe with lengths
df_lengths = pd.DataFrame({'length': lengths})

#print(df_lengths) ##this line is to check if dataframe is correct

# plotting histogram using plotnine with titles and interval length and colour of bars
p = ggplot(df_lengths, aes(x='length')) + geom_histogram(binwidth=1, alpha=0.8, color='black') + xlab("Length of text") + ylab("Frequency") + ggtitle("Histogram of Text Lengths")
# limiting the axis to the limits of our data
p = p + scale_x_continuous(limits=(0, 75)) + scale_y_continuous(limits=(0,250))
# making the background white
p = p + theme_classic(base_size=14, base_family='Arial')
# printing the histogram
print(p)



```
